```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("interactive_data.csv")
data
```

```python
np.unique(data["Intent"].to_numpy())
```

```python
total_gun_deaths = data[
    (data["Intent"]=="None selected")
    & (data["Gender"]=="None selected") 
    & (data["Age"]=="None selected")
    & (data["Race"]=="None selected")]["Deaths"].iloc[0]

suicide_deaths = data[(data["Intent"]=="Suicide")
        & (data["Gender"]=="None selected") 
        & (data["Age"]=="None selected")
        & (data["Race"]=="None selected")]["Deaths"].iloc[0]

suicide_deaths_proportion = int(100*suicide_deaths/total_gun_deaths)
print(f"The proportion of suicide deaths among all gun deaths is {suicide_deaths_proportion}%")
```

```python
suicide_data = data.loc[data["Intent"]=="Suicide"]
male_suicide = suicide_data.loc[suicide_data["Gender"]=="Male"]

male_suicide_deaths = male_suicide.loc[(male_suicide["Age"]=="None selected")
    & (male_suicide["Race"]=="None selected")]["Deaths"].iloc[0]
male_suicide_proportion = int(100*male_suicide_deaths/suicide_deaths)
print(f"The proportion of male persons among all suicide deathss is {male_suicide_proportion}%")
```

```python
homicide_deaths = data.loc[(data["Intent"]=="Homicide")
    & (data["Gender"]=="None selected")
    & (data["Age"]=="None selected")
    & (data["Race"]=="None selected")]["Deaths"].iloc[0]
homicide_proportion = int(100*homicide_deaths/total_gun_deaths)
print(f"The proportion of homicides among all gun deathss is {homicide_proportion}%")
```

```python
the_fourth_variable_total = data.loc[(data["Intent"]=="Homicide")
    & (data["Gender"]=="Male")
    & (data["Age"]=="15 - 34")
    & (data["Race"]=="None selected")]["Deaths"].iloc[0]

the_fourth_variable_black = data.loc[(data["Intent"]=="Homicide")
    & (data["Gender"]=="Male")
    & (data["Age"]=="15 - 34")
    & (data["Race"]=="Black")]["Deaths"].iloc[0]

the_fourth_proportion = int(100*the_fourth_variable_black/the_fourth_variable_total)

print(f"{the_fourth_proportion}% of homicide victims who are males in the age-group of 15--34 are black")
```

```python
woman_homicide = data.loc[(data["Intent"]=="Homicide")
    & (data["Gender"]=="Female")
    & (data["Race"]=="None selected")
    & (data["Age"]=="None selected")]["Deaths"].iloc[0]

woman_homicide_proportion = int(100*woman_homicide/homicide_deaths)

print(f"Women constitue only {woman_homicide_proportion}% of the total homicide victims")
```
